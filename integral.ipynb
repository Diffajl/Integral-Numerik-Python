```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as smp
```

```python
x = smp.symbols('x')
f = smp.sqrt(4-x**2)
hasil = smp.integrate(f,(x,-2,2))
print(hasil)
```

```python
def func(x):
    return np.sqrt(4-x**2)
```

```python
a = -2
b = 2
n = 5000

dx = (b-a)/(n-1)
x = np.linspace(a,b,n)

Sum = 0
for i in range(1,n-1):
    Sum += func(x[i])

hasil = 0.5*dx*(func(x[0])+
                2*Sum+func(x[-1]))
print(hasil)

nd = 1000
xp = np.linspace(a,b,nd)

plt.plot(xp,func(xp))

for i in range(n):
    plt.bar(x[i], func(x[i]),
    align='edge',width=0.000001,
    color='gray', edgecolor='red')

plt.fill_between(x,func(x),color='gray')

plt.show
```
